```python
import numpy as np
import pandas as pd

class CustLR():
    def __init__(self):
        self.m = None
        self.b = None
    def fit(self, X_train, Y_train):
        num = 0
        den = 0
        for i in range(len(X_train)):
            num += (Y_train[i] - np.mean(Y_train))*(X_train[i] - np.mean(X_train))
            den += (X_train[i] - np.mean(X_train))**2
        
        self.m = num/den
        self.b = np.mean(Y_train) - (self.m * np.mean(X_train))
        print(self.m)
        print(self.b)
    
    def predict(self, x):
        return self.m * x + self.b

df = pd.read_csv("./placement.csv")

df.head()

X = df.iloc[:, 0].values
y = df.iloc[:,1].values

from sklearn.model_selection import train_test_split



X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=2, test_size=0.2)
# print(X_train)
model = CustLR()

model.fit(X_train, Y_train)

Y_predict = model.predict(X_test) #Y^

def MAE():
    ans = 0
    sum = 0
    for i in range(len(Y_predict)):
        sum += abs(Y_predict[i] - Y_test[i])

    ans = sum / len(Y_predict)

    print("_________________________________-__-----_________________________________-__-----_________________________________-__-----")
    return ans
print(MAE())


def MSE():
    ans = 0
    sum = 0

    for i in range(len(Y_predict)):
        sum += (Y_predict[i] - Y_test[i])**2
    ans = sum / len(Y_predict)
    # print("akhfhkslialhblshifbsvb bsakrr skbi;saj")
    return ans

print(MSE())
print(MSE()**0.5)


def r2score():
    ssr = 0
    sst = 0
    meanY = np.mean(Y_test)

    for i in range(len(Y_test)):
        ssr = (Y_predict[i] - Y_test[i])**2
        sst = (Y_test[i] - meanY)**2
    return 1 - (ssr/sst)

print(r2score())

print("bkffj")
print(1-((1-r2score())*(len(Y_test)-1)/(len(Y_test)-1-1)))
```

```python
import numpy as np
import pandas as pd;
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

class CLR:
    def __init__(self):
        self.m = None
        self.b = None
    
    def fit(self, X_train, y_train):
        num = 0
        den = 0

        for i in range(X_train.shape[0]):
            num += (X_train[i] - np.mean(X_train)) * (y_train[i] - np.mean(y_train))
            den += (X_train[i] - np.mean(X_train))**2

        self.m = num/den
        self.b = np.mean(y_train) - self.m* np.mean(X_train)
        print(self.m)
        print(self.b)

    def predict(self, X_instance):
        return self.m*X_instance + self.b

data = pd.read_csv("./placement.csv")

data.head()

X = data.iloc[:, 0].values
y = data.iloc[:, 1].values

# type(np.array(data))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# print(np.shape(X_train))
plt.scatter(X_train, y_train, label='Train', alpha=0.6)
# plt.scatter(X_test, y_test, label='Test', alpha=0.6)
# plt.legend()
# plt.show()

# print(X_test)


model = CLR()

model.fit(X_train, y_train)

model.predict(7.73)
```
